```python
import os
import requests
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# === Configuration ===A
GITHUB_TOKEN = 'ghp_r1ILyJwgVd1mup0E5DagtJj2C9NEy'  
endpoint = "https://models.github.ai/inference"
model = "openai/gpt-4.1"
YOUR_YOUTUBE_API_KEY = "AIzaSyB3wotB3j29GopwAvzgsLnYgaQY"  

# === Azure Client Setup ===
client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(GITHUB_TOKEN),
)

# === YouTube Search Function ===
def get_youtube_videos(api_key: str, query: str, max_results: int = 2):
    search_url = "https://www.googleapis.com/youtube/v3/search"
    params = {
        "part": "snippet",
        "q": query,
        "key": api_key,
        "type": "video",
        "maxResults": max_results,
    }
    response = requests.get(search_url, params=params)
    if response.status_code != 200:
        return [f"Error fetching YouTube data for '{query}'."]

    data = response.json()
    video_links = []
    for item in data.get("items", []):
        video_id = item["id"]["videoId"]
        video_title = item["snippet"]["title"]
        video_url = f"https://www.youtube.com/watch?v={video_id}"
        video_links.append(f"- {video_title}\n  {video_url}")
    return video_links

# === User Input for Requirement ===
user_topic = input("Enter your learning topic (e.g., AI/ML, Data Science, etc.): ").strip()
days = 17
prompt = f"Generate a {days}-day detailed learning roadmap for: {user_topic}. Each day should have a specific learning task or concept."

# === Get Roadmap from Azure OpenAI ===
response = client.complete(
    messages=[
        SystemMessage("You are an expert educator and roadmap planner."),
        UserMessage(prompt),
    ],
    temperature=1,
    top_p=1,
    model=model
)

roadmap_text = response.choices[0].message.content
print("\n=== Initial Roadmap ===\n")
print(roadmap_text)

# === Enhance with YouTube Links ===
print("\n=== Roadmap with YouTube Recommendations ===\n")

import re

day_pattern = re.compile(r"(?i)(Day\s*\d+[:\-]?)\s*(.+)")
lines = roadmap_text.split('\n')
enhanced_roadmap = []

for line in lines:
    match = day_pattern.match(line.strip())
    if match:
        day_label = match.group(1).strip()
        task = match.group(2).strip()
        print(f"{day_label} {task}")
        videos = get_youtube_videos(YOUTUBE_API_KEY, task)
        for v in videos:
            print(f"   {v}")
        print()
        enhanced_roadmap.append(f"{day_label} {task}")
        enhanced_roadmap.extend([f"   {v}" for v in videos])
    else:
        enhanced_roadmap.append(line)

# === Optionally Save to File ===
with open(f"{user_topic.lower().replace(' ', '_')}_roadmap.txt", "w", encoding='utf-8') as f:
    f.write('\n'.join(enhanced_roadmap))

print("\n✅ Roadmap with YouTube resources saved.")
```
